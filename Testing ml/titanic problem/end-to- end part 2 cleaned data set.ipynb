```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline

titanic = pd.read_csv('C:/Users/Tharuka/Documents/Applied-Machine-Learning-Foundations/Ex_Files_Applied_Machine_Learning/Exercise Files/titanic_cleaned_dataset.csv')
titanic.head()
```

```python
#cabing is not missing  randomly
#Creating a boolean indicator for cabin to identify

titanic['cabin_ind'] = np.where(titanic['Cabin'].isnull(), 0, 1)
```

```python
#converting sex to numeric
gender_num = {'male':0,'female':1}
titanic['Sex'] = titanic['Sex'].map(gender_num)
```

```python
#drop unnecessary variables
titanic.drop(['Cabin','Embarked','Name','Ticket'], axis=1, inplace=True)
```
