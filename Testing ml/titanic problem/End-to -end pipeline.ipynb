```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline

titanic = pd.read_csv('C:/Users/Tharuka/Documents/Applied-Machine-Learning-Foundations/Ex_Files_Applied_Machine_Learning/Exercise Files/titanic.csv')
titanic.head()
```

```python
#to find missing data as a count in the data set
titanic.isnull().sum()
```

```python
#filling age values with a the mean value of age  as this ages are missing as random
titanic['Age'].fillna(titanic['Age'].mean(), inplace=True)
```

```python
#Combine SibSp & Parch 1


for i, col in enumerate(['SibSp','Parch']):
    plt.figure(i)
    sns.catplot(x=col, y='Survived', data=titanic, kind='point', aspect=2,)
```

```python
#Combine SibSp & Parch 2
titanic['Family_cnt'] =titanic['SibSp'] + titanic['Parch']
```

```python
#drop unnecessary variables

titanic.drop(['PassengerId', 'SibSp', 'Parch'], axis=1 , inplace=True)
```
